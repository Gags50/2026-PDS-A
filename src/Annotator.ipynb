```python
import matplotlib.pyplot as plt
import pandas as pd

metadata = pd.read_csv("./data/metadata.csv")
metadata = metadata[metadata["group_id"] == "A"]
metadata["hairiness"] = [None for x in range(len(metadata))]
metadata["marker"] = [None for x in range(len(metadata))]
positive = ["y", "yes", "1", "true"]
negative = ["n", "no", "0", "false"]

for patient in metadata.itertuples():
    img = patient.img_id
    plt.imshow(plt.imread(f"./data/imgs/{img}"))
    plt.show()
    metadata.loc[metadata["img_id"] == img, "hairiness"] = input("How hairy is the leisure? 0-3\n0: Not Hairy\n3: Very Hairy\n")
    answer = input("Does the image have a pen mark? Y/N\n").lower()
    if answer in positive:
        metadata.loc[metadata["img_id"] == img, "marker"] = 1
    elif answer in negative:
        metadata.loc[metadata["img_id"] == img, "marker"] = 0
    
metadata.to_csv("./data/annotated_metadata.csv")
```
